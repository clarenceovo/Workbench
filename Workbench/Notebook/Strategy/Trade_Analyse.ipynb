```python
#htx['Executed Time'] is GMT +8 , to UTC

import pandas as pd

# Parse the datetime strings and localize to Asia/Shanghai, then convert to UTC
htx['UTC_TIME'] = pd.to_datetime(htx['Executed Time']).dt.tz_localize('Asia/Shanghai').dt.tz_convert('UTC')
```

```python
htx
```

```python
bn['Date(UTC)'] = pd.to_datetime(bn['Date(UTC)'])
htx['UTC_TIME_NAIVE'] = htx['UTC_TIME'].dt.tz_convert('UTC').dt.tz_localize(None)
```

```python
merged_df = pd.merge(htx, bn, left_on='UTC_TIME_NAIVE', right_on='Date(UTC)', how='inner')

```

```python
merged_df.to_csv("BN_HTX.csv", index=False)
```
